# Colab 이란 무엇인가?

Colab, 또는 구글 코랩 (Google Colaboratory) 은 `클라우드 기반의 무료 Jupyter Notebook` 개발 환경입니다.

내부 기술 스택으로는 `Google Colab, Google 드라이브, 도커, 리눅스, Google 클라우드 기술 등`이 사용됩니다.

이 환경은 다음과 같은 장단점을 가지고 있습니다.

## 장점

1. 무료 사용 가능
    - Colab은 사용자에게 비용을 청구하지 않습니다. (다만, `고성능 GPU 사용을 원할 시 추가 비용`이 발생 가능!)

2. 쉬운 사용
    - 환경 설정과 사용 방법이 `간편`합니다.

3. 로컬 환경에 구애받지 않음
    - 사용자는 로컬 머신 사양에 의존하지 않고 `클라우드 서버를 이용`할 수 있습니다.

4. 협업 용이
    - 여러 사용자가 `동시에 수정`하고 `협업`할 수 있습니다.

5. 모바일 호환
    - 어디에서든 `인터넷 브라우저를 통해 접속하고 Python 스크립트를 작성하고 실행 가능`하며, `모바일에서도 작업`할 수 있습니다.

6. 우수한 성능
    - 일반적인 개인 PC보다 `빠른 성능을 제공`하며, `딥러닝 학습에 적합`합니다.

7. 고급 기능 제공
    - Colab은 Jupyter Notebook에 추가 기능을 제공하며, `Git과의 연동이 용이`합니다.

8. 오류 해결 도우미
    - 오류 발생 시 `Stack Overflow 검색 결과를 제공`하여 문제 해결을 도와줍니다.

## 단점

1. 세션 제한
    - 최대 세션 유지 시간은 12시간으로, 12시간 동안 아무 작업을 하지 않거나 `세션이 종료되면 데이터가 손실`됩니다. 그러나 소스 코드는 `.ipynb 확장자로 Google 드라이브에 안전하게 저장`되므로 걱정하지 않아도 됩니다.

2. 저장 용량 한정
    - Google 드라이브 개인 계정의 최대 용량은 15GB이며, 필요에 따라 `추가 저장 공간을 구입할 수 있습니다.`

## 결론

Google Colab은 `딥러닝 및 머신 러닝 프로젝트`를 쉽게 시작하고 협업할 수 있는 매우 유용한 도구 중 하나입니다.

- Reference
[Colaboratory에 오신 것을 환영합니다](https://colab.research.google.com/notebooks/welcome.ipynb)

# 3. Colab & Markdown 문법

[Markdown](https://www.markdownguide.org/getting-started/)은 텍스트 기반의 마크업언어로 2004년 존그루버에 의해 만들어졌으며 쉽게 쓰고 읽을 수 있으며 HTML로 변환이 가능합니다.

특수기호와 문자를 이용한 매우 간단한 구조의 문법을 사용하여 웹에서도 보다 빠르게 컨텐츠를 작성하고 보다 직관적으로 인식할 수 있습니다.

마크다운이 최근 각광받기 시작한 이유는 [Github](https://github.com) 덕분입니다.

깃헙의 저장소 (Repository)에 관한 정보를 기록하는 README.md는 깃헙을 사용하는 사람이라면 누구나 가장 먼저 접하게 되는 마크다운 문서입니다.

마크다운을 통해서 설치방법, 소스코드 설명, 이슈 등을 간단하게 기록하고 가독성을 높일 수 있다는 강점이 부각되면서 점점 여러 곳으로 퍼져가게 되었습니다.

또한, 마크다운 문법을 활용하면, Colab 에서 코드에 대한 개념을 `가독성이 좋도록 기록 및 정리` 할 수 있습니다.
